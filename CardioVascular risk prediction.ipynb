````python
# Creating Parameter Class 
class findz:
  def proportion(self,sample,hyp,size):
    return (sample - hyp)/math.sqrt(hyp*(1-hyp)/size)
  def mean(self,hyp,sample,size,std):
    return (sample - hyp)*math.sqrt(size)/std
  def varience(self,hyp,sample,size):
    return (size-1)*sample/hyp

variance = lambda x : sum([(i - np.mean(x))**2 for i in x])/(len(x)-1)
zcdf = lambda x: norm(0,1).cdf(x)
# Creating a function for getting P value
def p_value(z,tailed,t,hypothesis_number,df,col):
  if t!="true":
    z=zcdf(z)
    if tailed=='l':
      return z
    elif tailed == 'r':
      return 1-z
    elif tailed == 'd':
      if z>0.5:
        return 2*(1-z)
      else:
        return 2*z
    else:
      return np.nan
  else:
    z,p_value=stats.ttest_1samp(df[col],hypothesis_number)
    return p_value
    
  


# Conclusion about the P - Value
def conclusion(p):
  significance_level = 0.05
  if p>significance_level:
    return f"Failed to reject the Null Hypothesis for p = {p}."
  else:
    return f"Null Hypothesis rejected Successfully for p = {p}"

# Initializing the class
findz = findz()
````

### Hypothetical Statement - 1

1.Patients with heart-risk smokes average 10 ciggerete per day

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null hypothesis = H0 = 10

Alternative Hypothesis = H1 <10

#### 2. Perform an appropriate statistical test.

````python
from scipy import stats


# Subset the data to include only patients with heart risk
heart_risk_df = df[df['Output'] == 1]

# Subset the data to include patients who smoke and have heart risk
heart_risk_smokers_df = heart_risk_df[heart_risk_df['is_smoking'] == 1]

# Subset the data to include patients who don't smoke and have heart risk
heart_risk_non_smokers_df = heart_risk_df[heart_risk_df['is_smoking'] == 0]

# Calculate the mean number of cigarettes smoked per day by each group
smokers_mean = heart_risk_smokers_df['cigsPerDay'].mean()
non_smokers_mean = heart_risk_non_smokers_df['cigsPerDay'].mean()

# Calculate the t-statistic and p-value using a two-sample t-test
t_stat, p__value = stats.ttest_ind(heart_risk_smokers_df['cigsPerDay'], heart_risk_non_smokers_df['cigsPerDay'])

# Print the results
print("Smokers mean: ", smokers_mean)
print("Non-smokers mean: ", non_smokers_mean)
print("t-statistic: ", t_stat)
print("p-value: ",p__value)
````

````python
if p__value < 0.05:
  print(f'Null hypothesis rejected successfully and there is no difference between avg cigerrete of heart risk patient and non_heart-risk patient')
else:
  print(f'Failed to reject null hypothesis test and there is difference between avg cigerrete of heart risk patient and non_heart-risk patient')
````

````python
# Getting Z value
z = findz.mean(hypothesis_number,sample_mean,size,std)
# Getting P - Value
p = p_value(z=z,tailed='l',t="false",hypothesis_number=hypothesis_number,df=hypo_1,col="cigsPerDay")
# Getting Conclusion
print(conclusion(p))
````

Based on the analysis of the given dataset, it can be concluded that the risk of developing cardiovascular disease is significantly influenced by various factors such as age, gender, education level, body mass index, smoking habits, and blood pressure.

The dataset consisted of 4,253 observations and 16 features, which were initially preprocessed to handle missing values, outliers, and categorical variables. Exploratory data analysis was performed to get insights into the distribution of each feature, correlations among features, and class distribution.

The prevalence of smoking was found to be high among the patients, with an average of 9 cigarettes per day. The average systolic and diastolic blood pressure were found to be higher than the normal range, indicating that most of the patients had hypertension.

To predict the risk of cardiovascular disease, seven machine learning models were trained and evaluated using the preprocessed dataset. The models used for classification were Logistic Regression, K-Nearest Neighbor Classifier, Gaussian Naive Bayes Classifier, Support Vector Machine, XGB Classifier, Decision Tree Classifier, and Random Forest Classifier.

The performance of the models was evaluated using various metrics such as accuracy, precision, recall, F1-score, and ROC-AUC score. Among all the models, the Random Forest Classifier performed the best with an accuracy of 84.8%, followed by the XGB Classifier with an accuracy of 83.2%. The Logistic Regression model performed the worst with an accuracy of 67.1%.

The high accuracy of the Random Forest and XGB Classifier models can be attributed to their ability to handle non-linear relationships between features and their ability to capture feature interactions. These models also have the ability to handle large datasets and are less prone to overfitting.

The poor performance of the Logistic Regression model can be attributed to the assumption of linearity between the features and the outcome variable. This model is suitable for datasets with linear relationships, and the given dataset has non-linear relationships between features. The K-Nearest Neighbor Classifier and Decision Tree Classifier models also had relatively low accuracies, which may be due to their sensitivity to noise and their tendency to overfit on the training data.

In conclusion, the Random Forest Classifier and XGB Classifier models are the most suitable models for predicting the risk of cardiovascular disease, as they provide high accuracy and can handle non-linear relationships between features. However, further research and experimentation can be done to improve the performance of the models by optimizing the hyperparameters and selecting the best features for the model. Additionally, a larger and more diverse dataset can be used to improve the generalization of the models.
